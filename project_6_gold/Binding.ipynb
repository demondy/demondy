```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
```

```python
gold_am = pd.read_csv('./data/GOLDAM.csv', index_col=[0])
gold_pm = pd.read_csv('./data/GOLDPM.csv', index_col=[0])
gvzcls = pd.read_csv('./data/GVZCLS.csv', index_col=[0])
vxdxcls = pd.read_csv('./data/VXGDXCLS.csv', index_col=[0])
boa_ice = pd.read_csv('./data/BAML.csv', index_col=[0])
sp500 = pd.read_csv('./data/SP500.csv', index_col=[0])
```

```python
gold_am
```

```python
df = gold_am.join(gold_pm, how='outer').join(gvzcls, how='outer').join(vxdxcls, how='outer').join(boa_ice, how='outer').join(sp500, how='outer')
```

```python
df.head()
```

```python
df.dropna(subset=['SP500'], inplace=True)
```
